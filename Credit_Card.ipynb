***Minimum_payment is correlated with BALANCE the most. We can use the mode of BALANCE_to_MinPayment Ratio to predict missing Minimum_payment values.***

```python
df['Bal_MinPay']=df[df['MINIMUM_PAYMENTS'].notnull()]['BALANCE']/ df[df['MINIMUM_PAYMENTS'].notnull()]['MINIMUM_PAYMENTS']
Balance_MinPay = df['Bal_MinPay'].mean()
df.drop(columns='Bal_MinPay')
```

```python
# CREDIT_LIMIT  and MINIMUM_PAYMENTS has missing values so we need to fill.

df['CREDIT_LIMIT'].fillna(df['CREDIT_LIMIT'].median(),inplace=True)
df['CREDIT_LIMIT'].count()


#Using BALANCE_to_MinPayment Ratio to predit missing Minimum_payment values.
for i in range(len(df['MINIMUM_PAYMENTS'])):
    if pd.isnull(df['MINIMUM_PAYMENTS'][i]):
        df['MINIMUM_PAYMENTS'][i] = (df['BALANCE'][i] / Balance_MinPay) + 1  #In case Balance is 0, so add 1

df['MINIMUM_PAYMENTS'].count()
```

##### 3. Limit_usage (balance to credit limit ratio )
**Utilization Ratio-** Also called balance-to-limit ratio, is calculated by dividing the total of all your credit card balances by the total of all your credit card limits. High utilization can be an indicator of credit risk, so the lower your balance-to-limit ratio, the better.


```python
df['limit_usage']=df['BALANCE']/df['CREDIT_LIMIT']
df
```

### Applying PCA

```python
from sklearn.decomposition import PCA
```

```python
df_18cols_scaled.shape
```

```python
#We have 18 features so our n_component will be 18.
pca=PCA(n_components=18)
df_18cols_scaled_pca=pca.fit(df_18cols_scaled)
sum(df_18cols_scaled_pca.explained_variance_ratio_)
```

```python
var_ratio={}
for n in range(2,19):
    pca=PCA(n_components=n)
    df_18cols_scaled_pca =pca.fit(df_18cols_scaled)
    var_ratio[n]=sum(df_18cols_scaled_pca.explained_variance_ratio_)
    
var_ratio    
```

```python
pd.Series(var_ratio).plot()
```

***Since 10 components are explaining about 80% variance so we select 10 components***

```python
pca=PCA(n_components=7)
df_18cols_scaled_pca =pca.fit(df_18cols_scaled)
df_18cols_scaled_pca.explained_variance_
```

```python
np.sum(df_18cols_scaled_pca.explained_variance_)
```

```python
#Fitting the model and applying the dimensionality reduction with PCAs fit_transform.
reduced_pca= pd.DataFrame(pca.fit_transform(df_18cols_scaled), columns=['PC_' +str(i) for i in range(7)])
reduced_pca
```

***Now we have 7 variables after dimensionality reduction***

```python
pd.DataFrame(df_18cols_scaled_pca.components_.T, columns=['PC_' +str(i) for i in range(7)],index=df_18cols.columns.to_list())
```

#### Above data gave us eigen vector for each component

```python
# Factor Analysis : variance explained by each component- 
pd.Series(df_18cols_scaled_pca.explained_variance_ratio_,index=['PC_'+ str(i) for i in range(7)])
```

### Clustering

- To find K, we can start with a cluster range of 1 and 25

```python
from sklearn.cluster import KMeans
```

```python
cluster_range = range( 1, 25 )
cluster_errors = []

for num_clusters in cluster_range:
    clusters = KMeans( num_clusters )
    clusters.fit( reduced_pca )
    cluster_errors.append( clusters.inertia_ ) # clusters.inertia_ is basically cluster error here.
```

```python
clusters_df = pd.DataFrame( { "num_clusters":cluster_range, "cluster_errors": cluster_errors } )

clusters_df[0:25]
```

```python
# Plotting Elbow
plt.figure(figsize=(12,6))
plt.plot( clusters_df.num_clusters, clusters_df.cluster_errors, marker = "o" )
```

***From above graph we will find elbow range to be 5,6,7***

### Silhouette Coefficient

```python
from sklearn import metrics
```

```python
# Calculate Silhouette Coefficient for K=3 through K=12
k_range = range(5, 8)
scores = []
for k in k_range:
    km = KMeans(n_clusters=k, random_state=1)
    km.fit(reduced_pca)
    scores.append(metrics.silhouette_score(reduced_pca, km.labels_))
    
scores    
```

```python
# plot the results
plt.plot(k_range, scores)
plt.xlabel('Number of clusters')
plt.ylabel('Silhouette Coefficient')
plt.grid(True)
```

***K= 7 has the best Silhouette score***

```python
km = KMeans(n_clusters=7, random_state=1)
km.fit(reduced_pca)

#Plot
color_map={0:'r',1:'b',2:'g',3:'y',4:'c', 5:'k',6:'m'}
label_color=[color_map[i] for i in km.labels_]
plt.figure(figsize=(7,7))
plt.scatter(reduced_pca.loc[:,'PC_0'],reduced_pca.loc[:,'PC_1'],c=label_color,cmap='Spectral',alpha=0.1)
```

```python
df_pair_plot=pd.DataFrame(reduced_pca)
```

```python
df_pair_plot['Cluster']=km.labels_ #Add cluster column in the data frame
```

```python
# Conactenating labels found through Kmeans with data 
cluster_df =pd.concat([df_18cols ,pd.Series(km.labels_ ,name='Cluster')],axis=1)
cluster_df.head()
```

```python
# Mean value gives a good indication of the distribution of data. So we are finding mean value for each variable for each cluster
cluster = cluster_df.groupby('Cluster').mean()
cluster
```

```python
column_to_plot = ['MONTHLY_CASH_ADVANCE','MONTHLY_AVG_PURCHASES','limit_usage','payment_minpay','Installment','Oneoff']
column_to_plot_ = ['CASH_ADVANCE_FREQUENCY','PURCHASES_FREQUENCY','ONEOFF_PURCHASES_FREQUENCY','PURCHASES_INSTALLMENTS_FREQUENCY']
ax=cluster[column_to_plot].plot(kind='bar',figsize=(15,10))
plt.xlabel("Cluster")
plt.title("Insights")
ax.set(yscale="log")

plt.xticks(cluster.index, ('Cl-0', 'Cl-1', 'Cl-2', 'Cl-3', 'Cl-4', 'Cl-5', 'Cl-6'))
```
